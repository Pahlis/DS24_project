```python
import pandas as pd
import matplotlib.pyplot as plt

# Löser in data
df = pd.read_csv('C:/Users/lisap/Documents/Skolan/09_Projekt/npf-news/data/raw/scraped_articles_20251012_1921.csv')

# Översikt
print(f"Antal artiklar: {len(df)}")
print(f"Kolumner: {df.columns.tolist()}")
print(f"\nFörsta artikeln:")
print(df.iloc[0])

# Kollar datakvalitet
print(f"\nSaknade titlar: {df['title'].isna().sum()}")
print(f"Saknad text: {df['content'].isna().sum()}")

# Textlängd
df['text_length'] = df['content'].str.len()
print(f"\nGenomsnittlig textlängd: {df['text_length'].mean():.0f} tecken")

# Visualisera
df['text_length'].hist(bins=20)
plt.title('Fördelning av textlängd')
plt.show()
```
